# DBSCAN

This notebook demonstrates how to use the `DBSCAN` module from the `rice2025.unsupervised_learning` library.  

## Setup
Import necessary modules and load data. For this example, we will import the Two Moons dataset.

The Two Moons dataset is a synthetic dataset commonly used to evaluate clustering algorithms on non-linearly separable data.
- Points: Each point belongs to one of two interlocking moon-shaped clusters.
- Structure: The two moons form curved, semi-circular shapes positioned next to each other.

```python
# import library
from rice2025.unsupervised_learning import dbscan
import numpy as np
import matplotlib.pyplot as plt

# load dataset
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.08, random_state=42)
print("Dataset shape:", X.shape)
```

## Visualize the Clusters
We can visualize the data before running the algorithm.

```python
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], s=20, c='gray')
plt.title("Two Moons Dataset (Unlabeled)")
plt.show()
```

## Initialize and Fit DBSCAN Model

We use the `DBSCAN` class, which clusters points based on density. The class supports the following parameters:
- `eps`: The radius around a point used to determine which neighbors are close.
- `min_samples`: The minimum number of points needed within an eps-radius for a point to be considered a core point.

```python
# initialize model
model = dbscan.DBSCAN(eps=.2, min_samples=5)

# fit and get community labels
labels = model.fit_predict(X)
print(labels[:20]) # first 20
```

## Visualize Clustering Results

Each cluster will be plotted in a different color.  

```python
unique_labels = np.unique(labels)
print("Detected clusters:", unique_labels)

colors = []
for c in labels:
    if c == -1:
        colors.append("black")
    else:
        colors.append(plt.cm.tab20(c))

plt.figure(figsize=(7, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors, s=25)
plt.title("DBSCAN Cluster Results")
plt.show()
```
