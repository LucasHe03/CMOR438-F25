# Community Detection

This notebook demonstrates how to use the `Community Detection` module from the `rice2025.supervised_learning` library.  

## Setup
Import necessary modules and load data. For this example, we will import the Zachary Karate Club network.

The Zachary Karate Club network represents social interactions among 34 members of a university karate club.

- Nodes: Each node represents a club member (34 nodes total).
- Edges: An edge between two nodes indicates that those members interacted outside the club.
- Properties:
    - Undirected 
    - Unweighted

```python
# import library
from rice2025.unsupervised_learning import community_detection
import rice2025.utilities as util
import matplotlib.pyplot as plt
import numpy as np

# load dataset
import networkx as nx
G = nx.karate_club_graph()
adj_matrix = nx.to_numpy_array(G)
print(f"Adjacency matrix shape: {adj_matrix.shape}")
```

## Visualize the Network
We can visualize the graph before detecting communities.

```python
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title("Zachary's Karate Club Network")
plt.show()
```

## Initialize and Fit Community Detection Model

We use the `CommunityDetection` class, which labels nodes based on connected components.

```python
# initialize model
model = community_detection.CommunityDetection()

# fit and get community labels
labels = model.fit_predict(adj_matrix)
print("Community labels for each node:")
print(labels)
```

## Visualize Communities

We can color the nodes according to their detected community.

```python
unique_labels = np.unique(labels)
colors = [plt.cm.tab20(l / len(unique_labels)) for l in labels]

plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color=colors, edge_color='gray')
plt.title("Detected Communities in Karate Club Network")
plt.show()
```
