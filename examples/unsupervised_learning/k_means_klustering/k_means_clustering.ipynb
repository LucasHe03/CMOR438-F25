# K-Means Clustering

This notebook demonstrates how to use the `KMeans` module from the `rice2025.unsupervised_learning` library.  

## Setup
Import necessary modules and load data. For this example, we will import the Two Moons dataset.

The Two Moons dataset is a synthetic dataset commonly used to evaluate clustering algorithms on non-linearly separable data.
- Points: Each point belongs to one of two interlocking moon-shaped clusters.
- Structure: The two moons form curved, semi-circular shapes positioned next to each other.

```python
# import library
from rice2025.unsupervised_learning import k_means_clustering
import numpy as np
import matplotlib.pyplot as plt

# load dataset
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.08, random_state=42)
print("Dataset shape:", X.shape)
```

## Visualize the Data
We can visualize the data before running the algorithm.

```python
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], s=20, c='gray')
plt.title("Two Moons Dataset (Unlabeled)")
plt.show()
```

## Initialize and Fit KMeans Model

We use the `KMeansClustering` class, which clusters points by minimizing distances within clusters. The class supports the following parameters:
- `k`: The number of centroids to use.
- `max_iter`: The minimum number of iterations to determine clusters.

```python
# initialize model
model = k_means_clustering.KMeansClustering(k=2, max_iter=100)

# fit and get centroids
labels = model.fit(X)
labels = model.labels

print(labels[:20])
```

## Visualize Clusters

Each cluster will be plotted in a different color.  

```python
unique_labels = np.unique(labels)
colors = [plt.cm.tab10(int(l)) for l in labels]

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors, s=30)
plt.title("K-Means Clustering on Two Moons Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Visualize Centroids
We can also visualize the locations of the final centroids produced by the model.

```python
centroids = model.centroids

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors, s=30)
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    s=300,
    marker='X',
    edgecolor='black',
    linewidth=1.5,
    label="Centroids"
)
plt.title("K-Means Cluster Centroids")
plt.legend()
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
